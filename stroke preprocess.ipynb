### create y_train and y_test using label

```python

train_list = glob.glob(r"\train\*")
y_train = []
for i in range(len(train_list)):
    img_path = str(train_list[i])
    img = Image.open(img_path)
    arr_img = np.asarray(img)
    label = img_path.split(".")[1]
    if label == "normal":
        label_bin = 0
    else:
        label_bin = 1
    
    y_train.append(label_bin)
    
```
